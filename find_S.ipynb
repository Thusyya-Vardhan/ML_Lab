```python
import numpy as np

# Training data: each row is [features..., label]
# 'P' = positive, 'N' = negative
dataset = [
    ['Sunny', 'Warm', 'Normal', 'Strong', 'Yes'],
    ['Sunny', 'Warm', 'High',   'Strong', 'Yes'],
    ['Rainy', 'Cold', 'High',   'Strong', 'No'],
    ['Sunny', 'Warm', 'High',   'Strong', 'Yes'],
]

def find_s(dataset):
    hypothesis = None

    for sample in dataset:
        features, label = sample[:-1], sample[-1]
        
        if label == 'No':
            continue  # ignore negative examples
        
        if hypothesis is None:
            hypothesis = features[:]  # initialize with first positive
        else:
            for i in range(len(hypothesis)):
                if hypothesis[i] != features[i]:
                    hypothesis[i] = '?'  # generalize

    return hypothesis

h = find_s(dataset)
print("Final Hypothesis:", h)
```

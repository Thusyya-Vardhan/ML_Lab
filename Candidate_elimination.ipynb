```python

def CandidateElimination(data):
    S = data[0][:-1]          # first positive example
    G = ["?"] * len(S)

    for x in data:
        attrs, label = x[:-1], x[-1]

        if label == "Yes":
            for i in range(len(S)):
                if S[i] != attrs[i]:
                    S[i] = "?"
        else:
            for i in range(len(S)):
                if S[i] != attrs[i]:
                    G[i] = S[i]

dataset = [
    ['Sunny', 'Warm', 'Normal', 'Strong', 'Yes'],
    ['Sunny', 'Warm', 'High',   'Strong', 'Yes'],
    ['Rainy', 'Cold', 'High',   'Strong', 'No'],
    ['Sunny', 'Warm', 'High',   'Strong', 'Yes'],
]

S, G = CandidateElimination(dataset)

print("Specific Hyposthesis:", S)
print("General Hypothesis:", G)
```
